```python
def oneDGradDescent(stopping_par, R, N, x0):
    
    criteria, iterations,new_val_list,function_value_list = 0,0,[],[]
    fx = x0**2
    while(criteria<=4):
        fx,dfx=x0**2,2*x0
        x1=x0-R*dfx
        iterations+=1
        if(round(x1,stopping_par) == round (x0,stopping_par)):
            criteria+=1
        elif(iterations==N):
            criteria=5
        else:
            criteria=0
        new_val_list.append(x1)
        function_value_list.append(fx)
        
        print("Iteration: ", iterations)
        print("x0:", x0)
        print("fx: ", fx)
        print("dfx: ", dfx)
        print("\n")
        x0=x1
        
    print("Total iterations:", iterations)
    return(new_val_list,function_value_list)
```

#### 1.3 Implement the 2D Version

```python
def f(x0):
    return(x0.T @ a @ x0)

def grad(f):
    return((a.T + a) @ f)
```

```python
def twoDGradDescent(stopping_par,R,N,x0):
    
    criteria,iterations,x0x,x0y=0,0,float(x0[0]),float(x0[1])
    X_list=[x0x]
    Y_list=[x0y] 
    
    while(criteria<60):
        function=f(x0)
        derivative=grad(x0)
        x1=x0-R*derivative
        if(round(float(x1[0]), stopping_par) == round(float(x0[0]), stopping_par) 
           and round(float(x1[1]), stopping_par) == round(float(x0[1]), stopping_par)):
            criteria+=1
        elif(iterations==N):
            criteria=60
        else:
            criteria=0
        X_list.append(float(x1[0]))
        Y_list.append(float(x1[1]))
        x0=x1
        iterations+=1
        
    print('Convergence Point: ',x1)
    print('No. of iterations',iterations)
    print("Function value (f_x) is:", function)
    print("Gradient value (d_f_x) is:", derivative)
    print("Input value (x_0) is:", x0)
    
    return (X_list,Y_list)
```

```python
A = np.matrix('1,1;1,1')

np.linalg.det(A)
```

As the determinant here is zero, we can prove that the matrix is singular

```python
def condition_num(alpha, A):
    c = A+alpha*np.eye(2)
    eigenv = np.linalg.eig(c)
    condition_number=np.linalg.cond(c)
    print(eigenv, condition_number)
```
