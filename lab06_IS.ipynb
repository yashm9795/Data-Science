```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.svm import SVC
# support vector classifier
import sklearn.datasets as ds
```

```python
## Load iris data
iris_data = ds.load_iris()
i1 = 1
i2 = 3
X = iris_data['data'][:,(i1, i2)]
# select 2st and 4nd feature
y = iris_data['target']
print("features are :\n", X[:5])
print("target is :\n", y[:5])
```

```python
## plot your features here, colored with target
plt.figure(figsize=(10,9))
f0 = X[:,0]
f1 = X[:,1]
plt.scatter(f0, f1, c=y, alpha=0.5, s=100)
```

```python
## Train. 
model = SVC().fit(X, y)
```

```python
## Predict on grid
nGrid = 40
range1 = np.linspace(X[:,0].min(), X[:,0].max(), nGrid)
range2 = np.linspace(X[:,1].min(), X[:,1].max(), nGrid)
(x02d, x12d) = np.meshgrid(range1, range2)
X0 = x02d.ravel()
X1 = x12d.ravel()
Xp = np.column_stack((X0, X1))
# xp is a new X matrix using the points on regular grid.
# now predict the results using that matrix
# ... do it here
yp = model.predict(Xp)
yp = yp.reshape((len(X02d), len(X12d)))
```

```python
## example code for a nice image
plt.imshow(yp, alpha=0.3, extent=[X0.min(), X0.max(), X1.min(), X1.max()],
           origin='lower')
plt.contour(X02d, X12d, yp, levels=[0.5], colors='black')
plt.contour(X02d, X12d, yp, levels=[1.5], colors='black')
# also add the observed values as dots here

plt.scatter(X[:,0], X[:,1], c=y)
```
