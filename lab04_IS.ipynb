```python
## what you need to import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.neighbors import KNeighborsClassifier
import sklearn.datasets as ds
```

```python
## Load iris data, create x and y as different features
iris = ds.load_iris()
# first five rows
iris['data'][:5,:]
```

```python
x_first_two_cols = iris['data'][:,:2]
x_last_two_cols = iris['data'][:,2:4]
x_middle_two_cols = iris['data'][:,1:3]
# select 1st and 2nd feature, play with more features here
y = iris['target']
```

```python
## Train.  Note: no real training with NN
## use 5 neighborsfrom sklearn.neighbors import KNeighborsClassifier
X = x_first_two_cols
V = np.cov(X.T)
n = 5
p = 1 #Manhattan distance
mNN = KNeighborsClassifier(n_neighbors=5,p=1).fit(X,y)
```

```python
# scatterplot x, y with color defined by z
# alpha is transparency, s is size

xmin, xmax = X[:, 0].min() - 1, X[:, 0].max() + 1
ymin, ymax = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(xmin, xmax, 0.1),
                     np.arange(ymin, ymax, 0.1))

```

```python
from matplotlib.colors import ListedColormap

light_col = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
dark_col = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])


Z = mNN.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(figsize=(16,9))
plt.pcolormesh(xx, yy, Z, cmap=light_col)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=dark_col,
            edgecolor='k', s=45)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max()) 
plt.title("3-Class classification k = %i"
          % (n))
plt.show()
```
