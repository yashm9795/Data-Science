```python
#importing libraries

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

Y=np.array([[1,0],[0,1]])
x=np.array([[1],[2]])

def f1(x):
    a = x.T @ Y @ x
    return(a)



def f2(x):
    a = 2*Y@x
    return(a)


```

```python


def plot(x):
    e1 = np.linspace(-10,10,30)
    e2 = np.linspace(-10,10,30)
    
    (m1, m2) = np.meshgrid(e1, e2)
    
    x1 = m1.ravel()
    x2 = m2.ravel()
    

    Z = np.empty_like(m1)
    g1 = np.empty_like(m1)
    g2 = np.empty_like(m1)
    
    for i in range(30):
        for j in range(30):
            x = np.array((m1[i,j], m2[i,j]))
            Z[i,j] = f1(x)
            g = f2(x)
            g1[i,j] = g[0]
            g2[i,j] = g[1]

    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_aspect('equal')

    cs = ax.contour(e1, e2, Z)
 
    ax.clabel(cs, inline=1)

    ax.quiver(e1, e2, g1, g2)
```

```python
Y=np.array([[1,0.1],[0.1,1]])
plot(x)
```

```python
Y=np.array([[25,10],[25,10]])
C=np.linalg.eig(Y)
plot(x)
print(C)
```
