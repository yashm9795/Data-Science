```python
#GERMAN TANK PROBLEM
import random

largest=[]
count=0
for i in range(0,1000):
    tank=[]
    for i in range(3):
        tank.append(random.randint(1,101))
    
    tank.sort()
   
    largest.append(tank[2])


for eachelement in largest:
    if eachelement <= 71:
        count+=1
        

prob=count/1000
print(prob)
    
```
